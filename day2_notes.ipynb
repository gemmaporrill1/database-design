## Mongo DB
-----
- databases
- tables -> collections
- rows -> documents
- advantages:
1.
    - joins not required
    - no normalisation
        - nested documents
        - array of documents
- store data as you need it
2.
- flexible schema -> shape
    - can have:
        - different keys
        - extra keys
        - normal/ consistent 
- defines schema automatically
- doesn't need declaration of data types | reads automatically what data is provided
3.
- convert mongo data into BSON -> Binary JSON
    - optimized way
    - less space

## Comparison Operators

### Greater than 8
```
db.collection.find({
  rating: {
    "$gt": 8
  }
})
```

### Less than 8
```
db.collection.find({
  rating: {
    "$lt": 8
  }
})
```

### Less than and including 8
```
db.collection.find({
  rating: {
    "$lte": 8
  }
})
```

### Greater than and including 8
```
db.collection.find({
  rating: {
    "$gte": 8
  }
})
```

### All the movies which rating 8.4, 7, 8.1
```
db.collection.find({
  rating: {
    $in: [
      7,
      8.4,
      8.1
    ]
  }
})
```

### Not including the movies which rating 8.4, 7, 8.1
```
db.collection.find({
  rating: {
    $nin: [
      7,
      8.4,
      8.1
    ]
  }
})
```

## SQL comparison queries
### ID equal to 100
```
SELECT * from movies
WHERE id = 100
```

### Equal to 8
```
SELECT * from movies
WHERE rating = 8
```

### Greater than 8
```
SELECT * from movies
WHERE rating > 8
```

### Less than 8
```
SELECT * from movies
WHERE rating < 8
```

### Less than and including 8
```
SELECT * from movies
WHERE rating =< 8
```

### Greater than and including 8
```
SELECT * from movies
WHERE rating >= 8
```

### All the movies which rating 8.4, 7, 8.1
```
SELECT * FROM movies
WHERE rating IN (7, 8.4, 8.1);
```
